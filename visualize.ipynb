```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('output.dat', header=None, delim_whitespace=True)
df.columns = ['X', 'Y', 'U', 'V', 'P']

X = df['X'].drop_duplicates().to_numpy()
Y = df['Y'].drop_duplicates().to_numpy()
X = np.sort(X)
Y = np.sort(Y)

# Matrix, shape = (Y, X)
X_mat = np.zeros((len(X), len(Y)))
Y_mat = np.zeros((len(X), len(Y)))
U_mat = np.zeros((len(X), len(Y)))
V_mat = np.zeros((len(X), len(Y)))
P_mat = np.zeros((len(X), len(Y)))

for i, x_i in enumerate(X):
    X_mat[:, i] = x_i
    Y_mat[:, i] = Y    
    U_mat[:, i] = df[df['X'] == x_i]['U'].to_numpy()
    V_mat[:, i] = df[df['X'] == x_i]['V'].to_numpy()
    P_mat[:, i] = df[df['X'] == x_i]['P'].to_numpy()
```

# Second iteration

```python
fig, ax = plt.subplots()
c = ax.contourf(X_mat, Y_mat, U_mat, 100, cmap="jet")
fig.colorbar(c)
```

```python
fig, ax = plt.subplots()
c = ax.contourf(X_mat, Y_mat, V_mat, 100, cmap="jet")
fig.colorbar(c)
```

```python
fig, ax = plt.subplots()
c = ax.contourf(X_mat, Y_mat, P_mat, 100, cmap="jet")
fig.colorbar(c)
```

```python
plt.quiver(X_mat, Y_mat, U_mat, V_mat)
```
